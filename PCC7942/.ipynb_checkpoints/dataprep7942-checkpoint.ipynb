```python
import pandas as pd
```

```python
mRNA = pd.read_csv('PCC7942_mRNA.csv')
mRNA
```

```python
tss = pd.read_csv('PCC7942_TSSites.csv')
tss
```

```python
pcc7942 = pd.merge(tss, mRNA, on = 'Gene', how = 'inner')
pcc7942
```

```python
pcc7942 = pcc7942[['Scaffold', 'Transcription Start', 'Strand', 'Gene', 'mean']]
pcc7942
```

```python
pcc7942 = pcc7942[pcc7942['Scaffold'] == 'NC_007604']
```

```python
pcc7942
```

```python
pcc7942['Strand'].unique()
```

```python
trlist = []
comp_dic = {'A':'T','G':'C','C':'G','T':'A'}
prolen = 100
numpydf = pcc7942.to_numpy()
print(numpydf[:10])
```

```python
pccgene = []
pccseq = ''

data = open('promoter(7942)-txt.txt','r')
lines = data.readlines()
line = [line.strip() for line in lines]
pccseq = ''.join(line[0:])
data.close()
pccseq = pccseq.upper()
print(pccseq[:500])
```

```python
for i in numpydf:
    # chromosomal DNA
    if i[0] == 'NC_007604':
        #positive strands
        if i[2] == '+':
            # pccseq[i[2](TSS) - (prolen+1) (TSSë¡œë¶€í„° 100bp upstream) : TSS-1] - promoter sequence indexing, i[19]: RNA reads
            if (i[1]-(prolen+1)) > 0:
                trlist.append((pccseq[i[1]-(prolen+1):i[1]-1],i[4]))

        # negative strands
        else:
            compseq = ''
            seq = pccseq[i[1]:i[1]+prolen]
            for s in seq:
                compseq +=comp_dic[s]
            #ì—­ë°©í–¥ìœ¼ë¡œ complementary seq promoter ì¶”ì¶œ
            revcomp_seq = compseq[::-1]
            trlist.append((revcomp_seq, i[4]))
    else:
        break
```

```python
trlist
```

```python
final_list = []
```

```python
for i in range(len(trlist)):
    final_list.append(trlist[i][0])
```

```python
df = pd.DataFrame(final_list)
df
```

```python
df.to_csv('final.csv', index = False)
```
