```python
import numpy as np
import matplotlib.pyplot as plt

# Load the numpy files
y_test_fold0 = np.load('output/y_test_fold0.npy')
y_test_fold1 = np.load('output/y_test_fold1.npy')
y_test_fold2 = np.load('output/y_test_fold2.npy')
y_test_fold3 = np.load('output/y_test_fold3.npy')
y_test_fold4 = np.load('output/y_test_fold4.npy')

y_pred_fold0 = np.load('output/y_pred_fold0.npy')
y_pred_fold1 = np.load('output/y_pred_fold1.npy')
y_pred_fold2 = np.load('output/y_pred_fold2.npy')
y_pred_fold3 = np.load('output/y_pred_fold3.npy')
y_pred_fold4 = np.load('output/y_pred_fold4.npy')

# Concatenate the numpy arrays
y_test_all = np.concatenate((y_test_fold0, y_test_fold1, y_test_fold2, y_test_fold3, y_test_fold4), axis=0)
y_pred_all = np.concatenate((y_pred_fold0, y_pred_fold1, y_pred_fold2, y_pred_fold3, y_pred_fold4), axis=0)

correlation = np.corrcoef(y_test_all, y_pred_all)[0, 1]

print(f'Pearson correlation coefficient: {correlation}')

# Visualize y_test vs y_pred
plt.scatter(y_test_all, y_pred_all, color = 'blue', alpha = 0.5)
plt.xlim([0, 20])
plt.ylim([0, 20])
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Scatter plot of y_test vs y_pred')
plt.show()
```
